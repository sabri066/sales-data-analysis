```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Product': ['Laptop', 'Smart Watch' , 'Phone', 'Tablet', 'Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone', 'Tablet', 'Laptop'],
    'Price': [1200, 200, 800, 400, 1300, 850, 420, 1250, 820, 410, 1350],
    'Quantity': [5, 16, 10, 7, 3, 6, 8, 4, 9, 5, 2],
    'Region': ['East', 'West', 'East', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West']
}

df = pd.DataFrame(data)
df.index = range(1, len(df) + 1)
df
```

```python
df['Sales'] = df['Price'] * df['Quantity']
df
```

```python
print("Total Sales:", df['Sales'].sum())
print("Average Sales:", df['Sales'].mean())
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
plt.figure()
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
df = df.sort_values(by='Sales', ascending=False)
df
```

```python
product_sales = df.groupby('Product')['Sales'].sum()
print(product_sales)
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

```python
plt.figure()
product_sales.plot(kind='pie')
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Insight 1: Highest sales product
top_product = df.groupby('Product')['Sales'].sum().idxmax()
print("Top Selling Product:", top_product)

# Insight 2: Best region
top_region = df.groupby('Region')['Sales'].sum().idxmax()
print("Best Performing Region:", top_region)

# Insight 3: Average sales
avg_sales = df['Sales'].mean()
print("Average Sales:", avg_sales)
```
